
1.2.2 Principal Component Analysis, more commonly known as PCA, can be cleverly used to better visualize high dimensional data. 

> a. (1 point) Explain what PCA is and when it can be used. Make sure to provide an in-depth explanation and note what the drawbacks and limitations are.

The definition in the book is as follows
'Principal Components Analysis is a linear algebra technique for continuous attributes that finds new attributes (principal components) that (1) are linear combinations of the original attributes, (2) are orthogonal (perpendicular = loodrecht) to each other, (3) and capture the maximum amount of variation in the data.'

The goal of PCA is to find a new projection of your n-dimensional data points onto a smaller m-dimensional point. The first dimension captures as much variability as possible. The second dimension is orthogonal to the first dimension and will capture as much of the remaining variability as possible. 

Using PCA has a few upsides
- You can visualize a complex data set with many dimensions into a lower dimensional space.
- You can eliminate the redundant features and reduce noise. This can in turn result in the reduction of computational resources needed.
- You could use techniques that normally wouldn't work well on high dimensional data sets because of the dimension reduction.
- It tends to identify the strongest patterns in the data.

PCA can however only be used on numeric data.

A drawback of using PCA is the fact that it relies on linear assumptions. It is focused on finding orthogonal projections in a dataset that contains the highest variance possible. However, if you have a dataset that has nonlinear relationships, like when your data is presented in a circle, PCA will fail to find a good representative direction. See the image below for an example of this.



Another thing is that PCA does always find orthogonal principal components. It does however happen that this is not the most meaningful direction. The image below shows how PCA captures two directions that aren't necessarily the best directions.



To apply PCA we must first mean center the data. Mean centering means that the mean value for an attribute is subtracted from that attribute. 

> b. (0.5 points) Why do we first need to mean center the data before applying PCA?

PCA is a regressional model without intercept. This means that the regressions line will inevitably go through the origin. PCA takes the biggest variance in the data, however if the data is not centered the regression line could go through the wrong direction and not be taking the biggest variance. See the below image for an example.



In practice, applying PCA comes down to a series of matrix operations, this is useful, as matrix operations can be applied with little effort and can be computed in relatively little time.

The given data **X**, must first be mean centered. This can be easily done by calculating a row vector **** containing the mean values of each attribute and then subtracting it from **X** (**Y** = **X** - **1**, where **1** is a column vector of size **N** x 1, with N indicating the number of observations in the original data).

Then, the Singular Value Decomposition, or SVD, of **Y**, the mean centered data, can be calculated. **Y** = **USV**<sup>T</sup>. In practice, this is often done using the numpy.linalg.svd() function.

Using SVD on **Y** yields a series of vectors which can be used to project the data onto specific Principal Components (PCs). The entire dataset can be projected onto the Principal Components by multiplying **Z** = **Y\*V**, where **Z** indicates the projected data. Alternatively, one could project onto just a subset of all the PCs by indexing in the multiplication. For example: **Z** = **Y\*V[:,1]** would yield the projection of the data onto the first PC.

> c. (1 point)  Apply PCA (using the aforementioned method) on the Nanonose data and visualize a scatterplot of the projection of the data onto the first two PCs.

```python
X = np.array(df.as_matrix()) #Matrix with data

u = X.mean(axis=1) #Mean of the data rows
Y = X - u[:, np.newaxis] #Centered matrix

u, s, v = np.linalg.svd(Y) #Use SVD on Y

Z = Y * v[:,1] #Project on first two PCs

plt.plot(Z, 'ro')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid(True)
plt.title('Principal Component Analysis PC1 and PC2')
plt.show()
```

> d. (1 point) Alternatively, one could use EigenValue Decomposition, EVD, instead of SVD. What are the similarities and differences between SVD and EVD? Can both be applied in each case?

Both EVD and SVD are types of matrix factorization, EVD and SVD are both different matrix decompositions.

SVD works on m-by-n matrixes. SVD factorizes a linear operator into three simpler linear operators. The projection into an r-dimensional space. Element-wise multiplication with singular values and the transformation to the dimensional output space.

EVD can only be used on matrixes that are diagonalizable, so n-by-n matrixes. EVD is interested in vinding those vectors x which would not be rotated when transposing the matrix but who may scale in length with respect to the original vector.

The SVD and EVD are closely related because
- The left-singular vectors of M are eigenvectors of MM*
- The right-singular vectors of M are eigenvectors of M*M
- The non-zero singular values of M are the square roots of the non-zero eigenvalues of both M*M and MM*

You can't apply both in each case since EVD can only applied onto n-by-n matrixes where SVD can be applied on n-by-m matrixes.

PCA attempts to explain as much of the variance in data with as few PCs as possible. The variance explained by the m'th PC can be calculated using the following formula: 
$\rho_m = 1 - \frac{\|{Y} - {u}_m s_{mm} {v}_m^T\|^2_F}{\|{Y}\|^2_F} = \frac{s_{mm}^2}{\displaystyle \sum_{m'=1}^M s^2_{m'm'}}$

Which indicates that the variation $\rho$ for a given m can be calculated by dividing the squared singular value of component m by the sum of all squared singular values.

## 1.3 Similarity measures

We will use a subset of the data on wild faces described by Berg in 2005 transformed to a total
of 1000 gray scale images of size 40x40 pixels, we will attempt to find faces in the
data base that are the most similar to a given query face. To measure similarity we
will consider the following measures: SMC, Jaccard, Cosine, ExtendedJaccard, and
Correlation. These measures of similarity are described in *Introduction to Data Mining*, page 73-77 and are given by

\begin{eqnarray*}
\textrm{SMC}({x},{y}) & = & \frac{\textrm{Number of matching attribute values}}{\textrm{Number of attributes}} \\
\textrm{Jaccard}({x},{y}) & = & \frac{\textrm{Number of matching presences}}{\textrm{Number of attributes not involved in 00 matches}} \\
\textrm{Cosine}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\| \|{y}\|} \\
\textrm{ExtendedJaccard}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\|^2 + \|{y}\|^2 - {x}^T {y}} \\
\textrm{Correlation}({x},{y}) & = & \frac{\textrm{cov}({x},{y})}{\textrm{std}({x}) \textrm{std}({y})}
\end{eqnarray*}
where $\textrm{cov}({x},{y})$ denotes the covariance between ${x}$ and ${y}$ and $\textrm{std}({x})$ denotes the standard deviation of ${x}$.

Notice that the SMC and Jaccard similarity measures only are defined for binary
data, i.e., data that takes values in $\{0,1\}$. As the data we analyze is non-binary,
we will transform the data to be binary when calculating these two measures of
similarity by setting
$x_i = \left\{ \begin{array}{ll} 0 & \textrm{if~} x_i < \textrm{median}({x}) \\
                                1 & \textrm{otherwise.} \end{array} \right.$

The cosine formule can be explained by the fact that we multiply the X value with Y.T and devide this with the sqrt of X.T times the sqrt of Y.T. Because of this the X won't matter since output will be the same either way because of the Y.

The Extended Jaccard method substracts XYt at the end so it won't work for that formula.

The correlation won't work because it takes the zscore of both X and Y which will be different for a different X, resulting in a different result.

```python
import numpy as np
from scipy.stats import zscore

def similarity(X, Y, method):
    # type: (object, object, object) -> object
    X = np.mat(X)
    Y = np.mat(Y)
    N1, M = np.shape(X)
    N2, M = np.shape(Y)

    method = method[:3].lower()
    if method == 'ext':  # Extended Jaccard
        XYt = X * Y.T
        sim = XYt / (np.log(np.exp(sum(np.power(X.T, 2))).T * np.exp(sum(np.power(Y.T, 2)))) - XYt)
    elif method == 'cos':  # Cosine
        sim = (X * Y.T) / (np.sqrt(sum(np.power(X.T, 2))).T * np.sqrt(sum(np.power(Y.T, 2))))
    elif method == 'cor':  # Correlation
        X_ = zscore(X, axis=1, ddof=1)
        Y_ = zscore(Y, axis=1, ddof=1)
        sim = (X_ * Y_.T) / (M - 1)
    return sim

#First three formulas where a*x
#Cosine
if similarity(1, 2, 'cos') == similarity(2*1, 2, 'cos'):
    print('True for Cosine')
else:
    print('False for Cosine')

#Extended Jaccard
if similarity(1, 2, 'ext') == similarity(2*1, 2, 'ext'):
    print('True for Extended Jaccard')
else:
    print('False for Extended Jaccard')

#Correlation
if similarity(1, 2, 'cor') == similarity(2*1, 2, 'cor'):
    print('True for Correlation')
else:
    print('False for Correlation')

# Last three formulas where b+x
# Cosine
if similarity(1, 2, 'cos') == similarity(2+1, 2, 'cos'):
    print('True for Cosine')
else:
    print('False for Cosine')

# Extended Jaccard
if similarity(1, 2, 'ext') == similarity(2+1, 2, 'ext'):
    print('True for Extended Jaccard')
else:
    print('False for Extended Jaccard')

# Correlation
if similarity(1, 2, 'cor') == similarity(2+1, 2, 'cor'):
    print('True for Correlation')
else:
    print('False for Correlation')
```
